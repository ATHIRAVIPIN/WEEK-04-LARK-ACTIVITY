```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_excel('Telco-Customer-Churn.xlsx')
df
```

 1. Compare churn count with respect to gender

```python
sns.countplot(x='gender',data=df,hue='Churn')
```

 2. Find out how many female senior citizens there in the datase

```python
df[(df['gender']=='Female') &  (df['SeniorCitizen']==1.0)]['SeniorCitizen'].count()
```

```python
df2=df[(df['gender']=='Female') &  (df['SeniorCitizen']==1.0)]
sns.countplot(data=df2,x='SeniorCitizen') 
```

 3. Compare 'tenure' with 'Total Charges'

```python
sns.scatterplot(data=df,x='tenure',y='TotalCharges')
plt.title('Tenure Vs Total Charges')
plt.tight_layout()
```

 4. Find out which contract is preferred by the senior citizen.

```python
df3 = df[df['SeniorCitizen']==1.0]
sns.countplot(x='SeniorCitizen',data=df3, hue='Contract')
```
